# Modified page rank

```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
```

```python
node = 5
g = nx.erdos_renyi_graph(node,1,directed = True,seed=123)
nx.draw(g,with_labels = True)
```

```python
adjecy= nx.adjacency_matrix(g)
```

```python
Adjacency_matrix = adjecy.todense()
Adjacency_matrix
```

```python
row_sum = np.sum(Adjacency_matrix,axis = 0)
```

```python
row_sum
```

```python
transition_matrix = Adjacency_matrix/row_sum
```

```python
transition_matrix 
```

```python
Intial_prob =np.transpose( np.array([[1/4],[1/8],[1/4],[1/8],[1/4]]))
```

```python
Intial_prob2 =np.transpose( np.array([[1/5],[1/5],[1/5],[1/5],[1/5]]))
```

```python
np.sum(Intial_prob),np.sum(Intial_prob2)
```

```python
c1 = 0.8
c2 = 0.2
```

```python
P_next = c1*(np.matmul(Intial_prob,transition_matrix)) + c2*(Intial_prob2)
```

```python
P_next
```

# Threshold value

```python
e = 0.00005
```

```python
np.sum(P_next)
```

```python
np.sum(np.absolute(P_next - Intial_prob))
```

```python
while np.sum(np.absolute(P_next - Intial_prob))>e:
    Intial_prob = P_next
    print("updated prob %s"%Intial_prob)
    P_next = c1*(np.matmul(Intial_prob,transition_matrix)) + c2*(Intial_prob2)
    print('all page rank %s'%P_next)
```
